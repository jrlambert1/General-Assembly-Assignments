Exercise 1: Build a function called `r2_adj()` that will calculate $R^2_{adj}$ for a model. For inputs, you'll need `y_true`, `y_preds`, the average value of $y$, the number of variables $p$, and the number of observations $n$.

```python
def r2_adj(y_true, y_preds, y_mean, p, n):
    top = sum((y_true - y_preds) **2) * (1/(n - p - 1))
    bottom = sum((y_true - y_mean) **2) * (1/(n - 1))
    return 1 - (top/bottom)
```

```python
import pandas as pd
import numpy as np 
fires = './forestfires.csv'
forest = pd.read_csv(fires)
forest['tempf'] = forest['temp'] * 1.8 + 32
forest['hot and dry'] = forest['tempf'] * forest['DC']
forest['area_log'] = np.log(forest['area'] + 1)
```

```python
import statsmodels.api as sm
forest_new = forest.drop(['X', 'Y', 'month', 'day' , 'temp', 'RH', 'area'], axis=1)
forest_new = forest_new.dropna()
forest_new.head()
```

```python
dep = forest_new['area_log']
indep = forest_new.drop('area_log', axis=1)
indep = sm.add_constant(indep)
model = sm.OLS(dep, indep)
results = model.fit()
results.summary()
```

```python
y_pred_1 = results.predict(indep)
forest_new['y_mean'] = forest_new['area_log'].mean()
r2_adj(forest['area_log'], y_pred_1, forest_new['y_mean'], 8, len(y_pred_1))
```

```python
forest_new2 = forest_new.drop(['DMC', 'rain', 'y_mean'], axis=1)
forest_new2 = forest_new2.dropna()
dep = forest_new2['area_log']
indep = forest_new2.drop('area_log', axis=1)
indep = sm.add_constant(indep)
model = sm.OLS(dep, indep)
results = model.fit()
results.summary()
```

```python
y_pred_2 = results.predict(indep)
forest_new2['y_mean'] = forest_new2['area_log'].mean()
r2_adj(forest_new2['area_log'], y_pred_2, forest_new2['y_mean'], 6, len(y_pred_2))
```

Exercise 3: Suppose we build two models. Model 1 has predicted outputs `y_preds_1` and Model 2 has predicted outputs `y_preds_2`. We now have the question we'll always have:

![](https://spoonuniversity.com/wp-content/uploads/sites/87/2016/06/Rebecca-Black-Friday-Which-Seat-Can-I-Take-Meme-26.gif)

*(Okay, maybe which model should I pick?)*

Build a function:
- called `model_picker()`
- that accepts `y_true`, `y_preds_1`, and `y_preds_2` as inputs, and
- returns:
    - $R^2$, MSE, RMSE, MedAE, and MSLE for model 1,
    - $R^2$, MSE, RMSE, MedAE, and MSLE for model 2,
    - "Model 1 is the model to pick!" if all of Model 1's scores are better than Model 2,
    - "Model 2 is the model to pick!" if all of Model 2's scores are better than Model 1, or
    - "Neither model wins." if neither model is uniformly better than the other.


```python
def model_picker(y_true, y_preds_1, y_preds_2):
    top1 = sum((y_true - y_preds_1) **2)
    top2 = sum((y_true - y_preds_2) **2)
    bottom = sum((y_true - y_true.mean()) **2)
    r_2_1 = 1 - (top1/bottom)
    r_2_2 = 1 - (top2/bottom)
    MSE1 = (1/len(y_preds_1)) * top1
    MSE2 = (1/len(y_preds_2)) * top2
    RMSE1 = MSE1 ** 0.5
    RMSE2 = MSE2 ** 0.5
    MedAE1 = np.median(y_true - y_preds_1)
    MedAE2 = np.median(y_true - y_preds_2)
    MSLE1 = (1/len(y_preds_1)) * sum((np.log(y_preds_1 + 1) - np.log(y_true + 1)) **2)
    MSLE2 = (1/len(y_preds_2)) * sum((np.log(y_preds_2 + 1) - np.log(y_true + 1)) **2)
    lis1 = [r_2_1, MSE1, RMSE1, MedAE1, MSLE1]
    lis2 = [r_2_2, MSE2, RMSE2, MedAE2, MSLE2]
    check1 = [x for x, y in zip(lis1, lis2) if x > y]
    check2 = [y for x, y in zip(lis1, lis2) if x < y]
    if len(check1) == 5:
        print('Model 1 wins! Here: \n {} \n {}'.format(lis1, lis2))
        return
    elif len(check2) == 5:
        print('Model 2 wins! Here: \n {} \n {}'.format(lis1, lis2))
        return
    else:
        print('Neither Model wins', lis1, lis2)
        return
```

```python
model_picker(forest_new['area_log'], y_pred_1, y_pred_2)
```
